```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity

# ----------------------------
# Helper: compute PCA component matrix
# ----------------------------
def get_pca_matrix(df, n_features=38):
    df_numeric = df.select_dtypes(include=['number'])  # keep only numeric cols
    X = df_numeric.values
    pca = PCA(n_components=n_features)
    pca.fit(X)
    return pca.components_  # shape: (38, 38)

# ----------------------------
# Helper: cosine similarity between corresponding PCs
# ----------------------------
def cosine_similarity_vector(matrix1, matrix2):
    similarities = [
        cosine_similarity(matrix1[i].reshape(1, -1), matrix2[i].reshape(1, -1))[0, 0]
        for i in range(matrix1.shape[0])
    ]
    return similarities  # list of length 38

# ----------------------------
# Main pipeline
# ----------------------------

# 1. Load Dataset A and Dataset B
dfA = pd.read_csv("datasetA.csv")
dfB = pd.read_csv("datasetB.csv")

# 2. Ensure both have the same class labels
classes = sorted(set(dfA['class'].unique()) & set(dfB['class'].unique()))

results = {}

# 3. Loop through classes
for cls in classes:
    print(f"Processing class {cls}...")

    # Subset class-specific data
    subsetA = dfA[dfA['class'] == cls]
    subsetB = dfB[dfB['class'] == cls]

    # Compute PCA component matrices
    matrixA = get_pca_matrix(subsetA)
    matrixB = get_pca_matrix(subsetB)

    # Compute 38×1 similarity vector
    sim_vector = cosine_similarity_vector(matrixA, matrixB)

    # Store in dictionary
    results[cls] = sim_vector

# 4. Save results
similarity_df = pd.DataFrame(results)  # shape = (38, num_classes)
similarity_df.to_csv("classwise_cosine_similarity.csv", index=False)

print("✅ Saved classwise cosine similarity as classwise_cosine_similarity.csv")
```

```python
# ==========================
# STEP 1: Upload CSV files
# ==========================
from google.colab import files
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np

print("Upload Dataset A")
uploaded_A = files.upload()
print("Upload Dataset B")
uploaded_B = files.upload()

# Get filenames
fileA = list(uploaded_A.keys())[0]
fileB = list(uploaded_B.keys())[0]

dfA = pd.read_csv(fileA)
dfB = pd.read_csv(fileB)

print(f"Dataset A shape: {dfA.shape}")
print(f"Dataset B shape: {dfB.shape}")
print("Dataset A columns:", dfA.columns.tolist())
print("Dataset B columns:", dfB.columns.tolist())

# ==========================
# STEP 2: Detect label column
# ==========================
possible_labels = ["class", "Class", "label", "Label", "target", "Target", "sleep_stage"]

label_col = None
for col in possible_labels:
    if col in dfA.columns and col in dfB.columns:
        label_col = col
        break

if label_col is None:
    raise ValueError("❌ Could not detect the class/label column. "
                     "Please update `label_col` manually.")

print(f"✅ Using '{label_col}' as the class/label column")

# ==========================
# STEP 3: Helper functions
# ==========================
def get_top_eigenvectors(df, n_features=38, top_k=5):
    """Compute PCA and return top-k eigenvectors based on eigenvalue magnitude."""
    df_numeric = df.select_dtypes(include=['number'])
    X = df_numeric.values
    pca = PCA(n_components=n_features)
    pca.fit(X)

    # eigenvalues (explained variance)
    eigenvalues = pca.explained_variance_
    eigenvectors = pca.components_

    # sort indices by eigenvalue magnitude (descending)
    idx_sorted = np.argsort(eigenvalues)[::-1]

    # take top-k eigenvectors
    top_indices = idx_sorted[:top_k]
    top_eigenvectors = eigenvectors[top_indices]

    return top_eigenvectors

def correlation_between_eigenvectors(eigvecs_A, eigvecs_B):
    """Compute correlation between top-k eigenvectors of two datasets."""
    correlations = []
    for i in range(eigvecs_A.shape[0]):
        corr = np.corrcoef(eigvecs_A[i], eigvecs_B[i])[0, 1]
        correlations.append(corr)
    return correlations  # length = top_k

# ==========================
# STEP 4: Per-class Eigenvector Correlation
# ==========================
top_k = 5  # <-- change to 10 if you want top 10 eigenvectors
classes = sorted(set(dfA[label_col].unique()) & set(dfB[label_col].unique()))

results = {}

for cls in classes:
    print(f"Processing class {cls}...")

    # Subset each dataset by class
    subsetA = dfA[dfA[label_col] == cls]
    subsetB = dfB[dfB[label_col] == cls]

    # Get top-k eigenvectors
    topA = get_top_eigenvectors(subsetA, top_k=top_k)
    topB = get_top_eigenvectors(subsetB, top_k=top_k)

    # Compute correlations
    corr_vector = correlation_between_eigenvectors(topA, topB)

    # Store result
    results[cls] = corr_vector

# ==========================
# STEP 5: Save Results
# ==========================
correlation_df = pd.DataFrame(results, index=[f"PC{i+1}" for i in range(top_k)])
output_file = f"classwise_top{top_k}_eigenvector_correlations.csv"
correlation_df.to_csv(output_file)

print(f"✅ Saved results to {output_file}")
files.download(output_file)
```
