```python
# === INSTALL REQUIRED PACKAGES ===
!pip install -q xgboost catboost lightgbm scikit-learn

# === IMPORTS ===
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from google.colab import files

# === UPLOAD DATASET ===
print("📂 Please upload your 'top_35_features.csv' file...")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# === SPLIT FEATURES & TARGET ===
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
if y.dtype == 'object':
    y = LabelEncoder().fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# === BASE MODELS ===
xgb = XGBClassifier(
    n_estimators=1000,
    learning_rate=0.03,
    max_depth=8,
    subsample=0.9,
    colsample_bytree=0.9,
    use_label_encoder=False,
    eval_metric='mlogloss',
    random_state=42
)

cat = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.03,
    depth=8,
    colsample_bylevel=0.9,
    verbose=0,
    random_state=42
)

lgb = LGBMClassifier(
    n_estimators=1000,
    learning_rate=0.03,
    max_depth=8,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42
)

# === STACKING CLASSIFIER ===
stack_model = StackingClassifier(
    estimators=[
        ('xgb', xgb),
        ('cat', cat),
        ('lgb', lgb)
    ],
    final_estimator=LogisticRegression(max_iter=2000),
    passthrough=True,
    n_jobs=-1
)

# === TRAIN ===
stack_model.fit(X_train, y_train)
y_pred = stack_model.predict(X_test)

# === EVALUATION ===
print("\n🏆 Stacking Classifier Evaluation (All Features):")
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=4))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# === INSTALL REQUIRED PACKAGES ===
!pip install -q xgboost catboost lightgbm scikit-learn

# === IMPORTS ===
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier

# === UPLOAD DATA ===
from google.colab import files
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])

# === PREPARE FEATURES ===
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
if y.dtype == 'object':
    y = LabelEncoder().fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# === BASE MODELS ===
# === BASE MODELS ===
xgb = XGBClassifier(
    n_estimators=1000, learning_rate=0.03, max_depth=8,
    subsample=0.9, colsample_bytree=0.9,
    eval_metric='mlogloss', random_state=42
)

cat = CatBoostClassifier(
    iterations=1000, learning_rate=0.03, depth=8,
    colsample_bylevel=0.9,
    verbose=0, random_state=42
)

lgb = LGBMClassifier(
    n_estimators=1000, learning_rate=0.03, max_depth=8,
    subsample=0.9, colsample_bytree=0.9,
    random_state=42
)

base_models = [("xgb", xgb), ("cat", cat), ("lgb", lgb)]

# === OUT-OF-FOLD STACKING ===
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
meta_train = np.zeros((X_train.shape[0], len(base_models)))
meta_test = np.zeros((X_test.shape[0], len(base_models)))

for i, (name, model) in enumerate(base_models):
    oof_preds = np.zeros(X_train.shape[0])
    test_preds = np.zeros((X_test.shape[0], skf.n_splits))

    for fold, (tr_idx, val_idx) in enumerate(skf.split(X_train, y_train)):
        X_tr, X_val = X_train.iloc[tr_idx], X_train.iloc[val_idx]
        y_tr, y_val = y_train[tr_idx], y_train[val_idx]

        model.fit(X_tr, y_tr)
        oof_preds[val_idx] = model.predict_proba(X_val)[:, 1]
        test_preds[:, fold] = model.predict_proba(X_test)[:, 1]

    meta_train[:, i] = oof_preds
    meta_test[:, i] = test_preds.mean(axis=1)

# === META-LEARNER ===
meta_model = LGBMClassifier(
    n_estimators=500, learning_rate=0.05, max_depth=4,
    subsample=0.9, colsample_bytree=0.9,
    random_state=42
)
meta_model.fit(meta_train, y_train)
final_preds = meta_model.predict(meta_test)

# === EVALUATION ===
acc = accuracy_score(y_test, final_preds)
print(f"\n🏆 Final CV-Stacking Accuracy: {acc:.4f}")
```

```python
# === IMPORTS ===
!pip install -q interpret

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
from interpret.glassbox import ExplainableBoostingClassifier

# === Stability Selection for EBM Feature Ranking ===
def stable_ebm_ranking(X_train, y_train, n_iterations=10, top_n=None):
    feature_names = np.array(X_train.columns)
    importances_accum = {name: 0.0 for name in feature_names}

    for _ in range(n_iterations):
        # Bootstrap sample
        X_bs, y_bs = resample(X_train, y_train, stratify=y_train, random_state=None)

        ebm_tmp = ExplainableBoostingClassifier(
            interactions=0,  # no interactions for clean 1D importance
            learning_rate=0.05,
            max_bins=128,
            max_leaves=5,
            min_samples_leaf=2,
            max_rounds=500,
            random_state=None
        )
        ebm_tmp.fit(X_bs, y_bs)

        for term, score in zip(ebm_tmp.term_features_, ebm_tmp.term_scores_):
            if len(term) == 1:
                feature = feature_names[term[0]]
                if isinstance(score[0], (np.ndarray, list)):
                    total_score = np.sum([np.sum(np.abs(s)) for s in score])
                else:
                    total_score = np.sum(np.abs(score))
                importances_accum[feature] += total_score

    # Average importance
    for feat in importances_accum:
        importances_accum[feat] /= n_iterations

    importance_df = pd.DataFrame(list(importances_accum.items()), columns=['Feature', 'Importance'])
    importance_df = importance_df.sort_values(by='Importance', ascending=False).reset_index(drop=True)

    if top_n:
        return importance_df.head(top_n)
    return importance_df

# === Get Stable Feature Ranking ===
importance_df = stable_ebm_ranking(X_train, y_train, n_iterations=15)

# === Top-K Evaluation with Tuned EBM ===
feature_counts = list(range(5, min(50, X_train.shape[1] + 1), 5))
val_accuracies = []
test_accuracies = []

for k in feature_counts:
    top_k_features = importance_df['Feature'].iloc[:k].tolist()
    ebm_k = ExplainableBoostingClassifier(
        interactions=0,
        learning_rate=0.05,
        max_bins=128,
        max_leaves=5,
        min_samples_leaf=2,
        max_rounds=500,
        random_state=42
    )
    ebm_k.fit(X_train[top_k_features], y_train)

    val_acc = accuracy_score(y_val, ebm_k.predict(X_val[top_k_features]))
    test_acc = accuracy_score(y_test, ebm_k.predict(X_test[top_k_features]))

    val_accuracies.append(val_acc)
    test_accuracies.append(test_acc)
    print(f"Top {k} → Val Acc: {val_acc:.4f} | Test Acc: {test_acc:.4f}")

# === Plot Accuracies ===
plt.figure(figsize=(10, 6))
plt.plot(feature_counts, val_accuracies, 'o--', label="Validation Accuracy", color='blue')
plt.plot(feature_counts, test_accuracies, 'o-', label="Test Accuracy", color='skyblue')
plt.title("Stable EBM Accuracy vs Number of Top Features")
plt.xlabel("Number of Top Features (Stable Ranked)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.xticks(feature_counts)
plt.legend()
plt.tight_layout()
plt.show()

# === Best Accuracy ===
best_k = feature_counts[np.argmax(test_accuracies)]
print(f"\n🏆 Best accuracy with Top-{best_k} features: {max(test_accuracies):.4f}")
```
